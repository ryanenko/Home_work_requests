# Задача №1
## Кто самый умный супергерой?
Есть API по информации о супергероях с информацией по всем супергероям. 
Нужно определить кто самый умный(intelligence) из трех супергероев- Hulk, Captain America, Thanos.

```python

import requests


#Список со словарями для всех супергероев
file_list = [] 
url = ('https://akabab.github.io/superhero-api/api/all.json') 
response = requests.get(url)
file_list.extend(response.json())

#Функция для создания словаря, где ключ-имя супергероя, значение-нужная нам характеристика
hero_intel = {}
def hero_rat(name):
    for hero_dir in file_list:           
        hero_name = hero_dir['name']  
        hero_powerstats = hero_dir['powerstats']       
        if hero_name == name:            
            hero_com = {hero_name: hero_powerstats['intelligence']}
            hero_intel.update(hero_com)

#Заполняем словарь    
hero_rat('Hulk')
hero_rat('Captain America')
hero_rat('Thanos')

#Сортируем словарь по значению
sorted_tuples = sorted(hero_intel.items(), key=lambda item: item[1])
sorted_dict = {k: v for k, v in sorted_tuples}
hero_list = list(sorted_dict.keys())
#Вывод
print(f'Самый умный супергерой: {hero_list[-1]}' ) 
    
    
 



```

```python
import requests
from pprint import pprint
class YandexDisk:

    def __init__(self, token):
        self.token = token



 

    def _get_upload_link(self, disk_file_path):
        upload_url = "https://cloud-api.yandex.net/v1/disk/resources/upload"
        headers = self.get_headers()
        params = {"path": disk_file_path, "overwrite": "true"}
        response = requests.get(upload_url, headers=headers, params=params)
        pprint(response.json())
        return response.json()

    def upload_file_to_disk(self, disk_file_path, filename):
        resalt =  self._get_upload_link(disk_file_path=disk_file_path)
        href = resalt.get("href", "")
        response = requests.put(href, data=open(filename, 'rb'))
        response.raise_for_status()
        if response.status_code == 201:
            print("Success")



    
def file_token_ya():
    with open('C:/Users/Admin/Desktop/tokens/token_ya.txt', 'rt', encoding = 'utf-8') as f:
        d = f.read()
        return d

TOKEN = file_token_ya()
if  __name__ == '__main__':
    ya = YandexDisk(token=TOKEN)
    ya.upload_file_to_disk('disk_file_path', 'text.txt')



            
```
